# Avaliação Imobiliária
## Análise dos Dados
### Importações

```python
import matplotlib.pyplot as plt
import pandas            as pd
import seaborn           as sns
```

### Importação dos Dados

```python
amostra = pd.read_csv("amostra.csv", sep=';', decimal=",", thousands='.')
```

### Distribuição dos Valores por Variáveis

```python
fig, axs = plt.subplots(1, 5, figsize=(15, 5))
axs[0].boxplot(amostra['Area'])
axs[0].set_title('ÁREA')
axs[0].set_xlabel('Dados')
axs[0].set_ylabel('Valor')
axs[1].boxplot(amostra['Quartos'])
axs[1].set_title('QUARTOS')
axs[1].set_xlabel('Dados')
axs[1].set_ylabel('Valor')
axs[2].boxplot(amostra['Banheiros'])
axs[2].set_title('BANHEIROS')
axs[2].set_xlabel('Dados')
axs[2].set_ylabel('Valor')
axs[3].boxplot(amostra['Vagas'])
axs[3].set_title('VAGAS')
axs[3].set_xlabel('Dados')
axs[3].set_ylabel('Valor')
axs[4].boxplot(amostra['Unitario'])
axs[4].set_title('UNITÁRIO')
axs[4].set_xlabel('Dados')
axs[4].set_ylabel('Valor')
plt.tight_layout()
plt.show()
```

### Matriz de Relações entre Variáveis

```python
sns.pairplot(amostra)
```

### Matrix de Correlação

```python
matrix_corr = amostra.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(matrix_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
#plt.title('Matrix de Correlação')
plt.show()
```
